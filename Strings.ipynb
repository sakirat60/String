## String Formatting

Using f-strings

```python
name = "Derica"
age = 20
course = "Stylist"
```

```python
print(f"My name is {name}. I am {age} years old. I am learning {course}")
```

```python
print(f"My name is {name.upper()}. I am {age} years old. I am learning {course}")
```

```python
print(f"My name is {name.capitalize()}. I am {age} years old. I am learning {course}")
```
