# TASK 1- SPORTS PLAYER PERFORMANCE ANALYSIS

```python
#store the goal scores in a list
goal_scores = [12, 7, 9, 15,6,10]

print(goal_scores)
```

```python
type(goal_scores)
```

```python
#sorting the list of goal scores
goal_scores.sort()
```

```python
print(goal_scores)
```

```python
#calculate the total goal scored
total_goals = sum(goal_scores)
```

```python
print(total_goals)
```

```python
#highest goal scored
max(goal_scores)
```

```python
#lowest goal scored
min(goal_scores)
```

```python
#calculate the difference between the highest and lowestbgoal scored
difference = max(goal_scores) - min(goal_scores)
```

```python
print(difference)
```

```python
#to count the number of goal scores in the list
len(goal_scores)
```

```python
#calculate the average goal scored
average = total_goals / len(goal_scores)
```

```python
print(average)
```

```python
#Add a new score to the list
goal_scores.append(11)
```

```python
print(goal_scores)
```

```python
#Remove th lowest score from the list
goal_scores.remove(min(goal_scores))
```

```python
print(goal_scores)
```

### Why tuple is more suitable than a list for storing CPU temperature data.
